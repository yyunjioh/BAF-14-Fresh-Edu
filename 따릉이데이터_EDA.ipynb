# Rented Bike Count와 다른 변수 관계 알아보기

```python
data['Seasons'].value_counts()
```

```python
#plt 라이브러리
plt.bar(['Spring','Summer','Autumn','Winter'],data['Seasons'].value_counts().values-2000,color='red',alpha=0.5)
plt.title('Season barplot',fontsize=10)
plt.ylabel('count')
plt.show()

#sns 라이브러리
sns.countplot(data['Seasons'])
plt.title('Season barplot',fontsize=20)
```

```python
data['Rented Bike Count'].value_counts()
```

```python
sns.countplot(data['Rented Bike Count'])
plt.title('Bike Count barplot',fontsize=20)
plt.show()
```

```python
print(data[['Seasons','Rented Bike Count']].value_counts())
print(data.groupby('Seasons').mean()['Rented Bike Count'])
```

```python
data['Hour'].value_counts()
sns.countplot(data['Hour'])
plt.title('Hour barplot',fontsize=20)
plt.show()
print(data[['Hour','Rented Bike Count']].value_counts())
print(data.groupby('Hour').mean()['Rented Bike Count'])
```
